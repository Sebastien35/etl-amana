```python
from sqlalchemy import create_engine # noqa
import pandas as pd
import json

# crée un moteur SQLAlchemy pour MySQL
engine = create_engine("mysql+pymysql://user:password@localhost/mydatabase")


conn = engine.connect()

query = """
SELECT *
FROM ecommerce;
"""

df = pd.read_sql(query, conn)   
conn.close()
display(df.head())


# export json
data = df.to_dict(orient='records')
with open('ecommerce.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
    
# lire le fichier json
with open('ecommerce.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
display(data[:2])  # afficher les deux premiers enregistrements
```
